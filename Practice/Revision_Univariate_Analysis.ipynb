```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = "Practice_Data.csv"
df = pd.read_csv(file)

df.head(), df.info()
```

Frequency Distribution Table

```python
#Set Plot Style
sns.set_style("whitegrid")

#Create subplots for categorical variables
fig, axes = plt.subplots(1, 3, figsize=[18, 5])

#Create Frequency Distribution Table for "Response"
sns.countplot(x=df['Response'], ax=axes[0], palette="pastel")
axes[0].set_title("Frequency Distribution Table for 'Response'")

#Create Frequency Distribution Table for "Sex"
sns.countplot(x=df['Sex'], ax=axes[1], palette="pastel")
axes[1].set_title("Frequency Distribution Table for 'Sex'")

#Create Frequency Distribution Table for "Job"
sns.countplot(x=df['Job'], ax=axes[2], palette="pastel", order=df['Job'].value_counts().index)
axes[2].set_title("Frequency Distribution Table for 'Job'")
plt.xticks(rotation=90)

plt.tight_layout();
```

To enhance readability u can replace x by y

Create Histogram and Degree of Deviation

```python
#Create histograms for continuous variables
fig, axes = plt.subplots(figsize=[14, 5])

# Histogram for Age
sns.histplot(df['Age'], ax=axes[0], kde=True, bins=20, color="skyblue")
axes[0].set_title("Histogram for 'Age'")

# Histogram for charge_Monday
sns.histplot(df['charge_Monday'], ax=axes[1], kde=True, bins=20, color="salmon")
axes[1].set_title("Histogram for 'charge_Monday'")

plt.tight_layout();

# Calculate descriptive statistics
df[['Age', 'charge_Monday']].describe()
```
